# Création du réseau

```python
from tensorflow.keras import layers
from tensorflow.keras import Model
# Our input feature map is 150x150x3: 150x150 for the image pixels, and 3 for
# the three color channels: R, G, and B
img_input = layers.Input(shape=(150, 150, 3))

# First convolution extracts 16 filters that are 3x3
# Convolution is followed by max-pooling layer with a 2x2 window
x = layers.Conv2D(16, 3, activation='relu')(img_input)
x = layers.MaxPooling2D(2)(x)

# Second convolution extracts 32 filters that are 3x3
# Convolution is followed by max-pooling layer with a 2x2 window
x = layers.Conv2D(32, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)

# Third convolution extracts 64 filters that are 3x3
# Convolution is followed by max-pooling layer with a 2x2 window
x = layers.Conv2D(64, 3, activation='relu')(x)
x = layers.MaxPooling2D(2)(x)


# Flatten feature map to a 1-dim tensor so we can add fully connected layers
x = layers.Flatten()(x)

# Create a fully connected layer with ReLU activation and 512 hidden units
x = layers.Dense(512, activation='relu')(x)

# Create output layer with a single node and sigmoid activation
output = layers.Dense(250, activation='softmax')(x)

# Create model:
# input = input feature map
# output = input feature map + stacked convolution/maxpooling layers + fully 
# connected layer + sigmoid output layer
model = Model(img_input, output)

model.summary()
```

## Split du dataset (Train et validation)

```python
train_images_tup = []
for i in range(0,15000):
    train_images_tup.append(img_list_resized[i])
    
validation_images_tup = []
for i in range(15000,20985):
    validation_images_tup.append(img_list_resized[i])
```

```python
len(train_images_tup), len(validation_images_tup)
```

```python
train_images = []
train_label = []
for i in train_images_tup:
    train_images.append(i[0])
    train_label.append(i[1])
    
val_images = []
val_label = []
for i in validation_images_tup:
    val_images.append(i[0])
    val_label.append(i[1])
```

## Changement du typage pour TensorFlow v2.0.0

```python
IMG_SIZE = 150
train_images = np.array(train_images).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
val_images = np.array(val_images).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
```

```python
train_label = np.array(train_label) 
val_label = np.array(val_label) 
```

# Training

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

history = model.fit(train_images, train_label, epochs=10, 
                    validation_data=(val_images, val_label))
```

# Evaluation du modèle

```python
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
#plt.ylim([100, 1])
plt.ylim([0.5, 1])
plt.legend(loc='lower right')

test_loss, test_acc = model.evaluate(val_images,  val_label, verbose=2)
```
